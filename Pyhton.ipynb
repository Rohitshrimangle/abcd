```python
a="Rohit is Studying very hard"
v=["a","e","i","o","u"]
space=[" "]
s=0
count=0
c=0
for i in a.lower():
    if i in v:
        count=count+1
    elif i.isalpha():
        c=c+1
    elif i in space:
        s=s+1
print("Total count:",len(a))
print("Vowels:",count)
print("consonents",c)
print("space:",s)
```
