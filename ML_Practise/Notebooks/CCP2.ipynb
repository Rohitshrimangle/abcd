Data Cleaning

```python
df_data=df.copy()
```

```python
df_data.TotalCharges=pd.to_numeric(df_data.TotalCharges, errors='coerce')
df_data.isnull().sum()
```

```python
df_data.loc[df_data['TotalCharges'].isnull()==True]
```

```python
df_data.dropna(how ='any', inplace=True)
```

```python
print(df_data['tenure'].max())
```

```python
labels=['{0}-{1}'.format(i,i+11) for i in range(1,72,12)]

df_data['tenure_group']=pd.cut(df_data.tenure, range(1,80,12),right=False,labels=labels)
```

```python
df_data['tenure_group'].value_counts()
```

```python
df_data.drop(columns=['customerID','tenure'],axis=1,inplace=True)
df_data.head()
```

```python
for i, predictor in enumerate(df_data.drop(columns=['Churn','TotalCharges','MonthlyCharges'])):
    plt.figure(i)
    sns.countplot(data=df_data, x=predictor, hue='Churn')
```

```python
df_data['Churn']=np.where(df_data.Churn=='Yes',1,0)
```

```python
df_data.head()
```

```python
df_data_dummies=pd.get_dummies(df_data)
df_data_dummies.head()
```

```python
sns.lmplot(data=df_data_dummies, x='MonthlyCharges', y='TotalCharges',fit_reg=True)
```

```python
Mth=sns.kdeplot(df_data_dummies.MonthlyCharges[(df_data_dummies['Churn']==0)], color='Red', shade =True)
Mth=sns.kdeplot(df_data_dummies.MonthlyCharges[(df_data_dummies['Churn']==1)], ax=Mth, color='Blue', shade=True)
Mth.legend(['No Churn','Churn'], loc='upper right')
Mth.set_ylabel('Density')
Mth.set_xlabel('MonthlyCharges')
Mth.set_title('Monthly charges by churn')
```

```python
Tot= sns.kdeplot(df_data_dummies.TotalCharges[(df_data_dummies['Churn']==0)], color='Red', shade=True)
Tot= sns.kdeplot(df_data_dummies.TotalCharges[(df_data_dummies['Churn']==1)],ax=Tot, color='Blue', shade=True)
Tot.legend(['No Churn', 'Churn'],loc='upper right')
Tot.set_ylabel('Density')
Tot.set_xlabel('Total Charges')
Tot.set_title('Total charges by churn')
```

```python
new_df1_target0=df_data.loc[df_data['Churn']==0]
new_df1_target1=df_data.loc[df_data['Churn']==1]
```

```python
def uniplot(df,col, title,hue=None):

    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams['axes.labelsize']=20
    plt.rcParams['axes.titlesize']=22
    plt.rcParams['axes.titlepad']=30

    temp = pd.Series(data=hue)
    fig, ax = plt.subplots()
    width = len(df[col].unique())+7 +4*len(temp.unique())
    fig.set_size_inches(width, 8)
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data = df, x=col, order=df[col].value_counts().index, hue=hue,palette='bright')
    plt.show()
```

```python
uniplot(new_df1_target1, col='Partner', title='Distribution of gender for churned customers', hue='gender')
```

```python
uniplot(new_df1_target0,col='Partner',title='Distribution of Gender for Non Churned Customers',hue='gender')
```
