```python
a="Rohit is Studying very hard"
v=["a","e","i","o","u"]
space=[" "]
s=0
count=0
c=0
for i in a.lower():
    if i in v:
        count=count+1
    elif i.isalpha():
        c=c+1
    elif i in space:
        s=s+1
print("Total count:",len(a))
print("Vowels:",count)
print("consonents",c)
print("space:",s)
```

```python
a="Rohit is Studying Hard"
upper=0
lower=0
special=0
digit=0

for i in a:
    if i.isupper():
        upper=upper+1
    elif i.islower():
        lower=lower+1
    elif i.isdigit():
        digit=digit+1
    else:
        special=special+1
print("Count of Upper:",upper)
print("Count of Lower:",lower)
print("Count of digit:",digit)
print("Count of special:",special)
```
